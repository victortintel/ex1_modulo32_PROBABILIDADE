3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

- Calculando as probabilidades para Y2. Neste caso o exercício nos pede para fazermos 
uma distribuição de t-studdent com 5 graus de liberdade, 
A distribuição t-studdent é muito parecida com a 
distribuição normal, a diferença é que normalmente 
ela produz valores mais extremos.
(quanto maior o grau de liberdade mais parecido é com uma distribuição normal)

```python
# aqui o próprio exercício já fez esta definição. Neste ele faz um array que contém 200 valores entre -6 e 6
#O exercício já nos passou que a média é 0 e a variância é 1. 
# Calculando as probabilidades para Y1.Aqui é muito parecido com o exercício anterior(Lembre que o loc significa média e o scale é o desvio padrão)

Y=np.linspace(-6, 6, 200)
# Seu código começa aqui
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)
prob_Y2 = stats.t.cdf(Y, df=5)
```

```python
print("Probabilidades para Y1 (distribuição normal padrão):")

for i in range(len(Y)):

  print("Probabilidade de Y1 < {:.2f}: {:.4f}".format(Y[i], prob_Y1[i]))
```

```python
print("Probabilidades para Y2 (distribuição normal padrão):")

for i in range(len(Y)):

  print("Probabilidade de Y2 < {:.2f}: {:.4f}".format(Y[i], prob_Y2[i]))
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
Y = np.linspace(-6, 6, 200)
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)
prob_Y2 = stats.t.cdf(Y, df=5)

plt.figure(figsize=(10, 6))

plt.plot(Y, prob_Y1, label='Distribuição Normal Padrão (Y1)')

plt.plot(Y, prob_Y2, label='Distribuição t de Student (5 graus de liberdade) (Y2)')

plt.title('Probabilidades de Y1 e Y2 serem menores que valores em Y')

plt.xlabel('Valores de Y')

plt.ylabel('Probabilidade')

plt.legend()

plt.grid(True)

plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui
prob_Y1_ppf = stats.norm.ppf(Y, loc=0, scale=1)
prob_Y2_ppf = stats.t.ppf(Y, df=5)
```

```python
print("Probabilidades para Y1_ppf (distribuição normal padrão):")

for i in range(len(Y)):

  print("Probabilidade de Y1_ppf < {:.2f}: {:.4f}".format(Y[i], prob_Y1_ppf[i]))
```

```python
print("Probabilidades para Y2_ppf (distribuição normal padrão):")

for i in range(len(Y)):

  print("Probabilidade de Y2_ppf < {:.2f}: {:.4f}".format(Y[i], prob_Y2_ppf[i]))
```

6. Faça um gráfico com os dados do item anterior.

```python
Y=np.linspace(0, 1, 20)

prob_Y1 = stats.norm.ppf(Y, loc=0, scale=1)
prob_Y2 = stats.t.ppf(Y, df=5)

plt.figure(figsize=(10, 6))

plt.plot(Y, prob_Y1, label='Distribuição Normal Padrão (Y1)')

plt.plot(Y, prob_Y2, label='Distribuição t de Student (5 graus de liberdade) (Y2)')

plt.title('Probabilidades de Y1 e Y2 serem menores que valores em Y')

plt.xlabel('Valores de Y')

plt.ylabel('Probabilidade')

plt.legend()

plt.grid(True)

plt.show()
```
