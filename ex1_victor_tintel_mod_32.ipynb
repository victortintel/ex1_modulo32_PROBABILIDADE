3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

- Calculando as probabilidades para Y2. Neste caso o exerccio nos pede para fazermos 
uma distribuio de t-studdent com 5 graus de liberdade, 
A distribuio t-studdent  muito parecida com a 
distribuio normal, a diferena  que normalmente 
ela produz valores mais extremos.
(quanto maior o grau de liberdade mais parecido  com uma distribuio normal)

```python
# aqui o prprio exerccio j fez esta definio. Neste ele faz um array que contm 200 valores entre -6 e 6
#O exerccio j nos passou que a mdia  0 e a varincia  1. 
# Calculando as probabilidades para Y1.Aqui  muito parecido com o exerccio anterior(Lembre que o loc significa mdia e o scale  o desvio padro)

Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)
prob_Y2 = stats.t.cdf(Y, df=5)
```

```python
print("Probabilidades para Y1 (distribuio normal padro):")

for i in range(len(Y)):

  print("Probabilidade de Y1 < {:.2f}: {:.4f}".format(Y[i], prob_Y1[i]))
```

```python
print("Probabilidades para Y2 (distribuio normal padro):")

for i in range(len(Y)):

  print("Probabilidade de Y2 < {:.2f}: {:.4f}".format(Y[i], prob_Y2[i]))
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
Y = np.linspace(-6, 6, 200)
prob_Y1 = stats.norm.cdf(Y, loc=0, scale=1)
prob_Y2 = stats.t.cdf(Y, df=5)

plt.figure(figsize=(10, 6))

plt.plot(Y, prob_Y1, label='Distribuio Normal Padro (Y1)')

plt.plot(Y, prob_Y2, label='Distribuio t de Student (5 graus de liberdade) (Y2)')

plt.title('Probabilidades de Y1 e Y2 serem menores que valores em Y')

plt.xlabel('Valores de Y')

plt.ylabel('Probabilidade')

plt.legend()

plt.grid(True)

plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui
prob_Y1_ppf = stats.norm.ppf(Y, loc=0, scale=1)
prob_Y2_ppf = stats.t.ppf(Y, df=5)
```

```python
print("Probabilidades para Y1_ppf (distribuio normal padro):")

for i in range(len(Y)):

  print("Probabilidade de Y1_ppf < {:.2f}: {:.4f}".format(Y[i], prob_Y1_ppf[i]))
```

```python
print("Probabilidades para Y2_ppf (distribuio normal padro):")

for i in range(len(Y)):

  print("Probabilidade de Y2_ppf < {:.2f}: {:.4f}".format(Y[i], prob_Y2_ppf[i]))
```

6. Faa um grfico com os dados do item anterior.

```python
Y=np.linspace(0, 1, 20)

prob_Y1 = stats.norm.ppf(Y, loc=0, scale=1)
prob_Y2 = stats.t.ppf(Y, df=5)

plt.figure(figsize=(10, 6))

plt.plot(Y, prob_Y1, label='Distribuio Normal Padro (Y1)')

plt.plot(Y, prob_Y2, label='Distribuio t de Student (5 graus de liberdade) (Y2)')

plt.title('Probabilidades de Y1 e Y2 serem menores que valores em Y')

plt.xlabel('Valores de Y')

plt.ylabel('Probabilidade')

plt.legend()

plt.grid(True)

plt.show()
```
